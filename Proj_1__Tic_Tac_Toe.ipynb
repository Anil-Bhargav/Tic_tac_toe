```python
from IPython.display import clear_output

def print_board(board):
    for item in board:
        for i in xrange(len(item)):
            print ' -----',
        print ''
        print ' | ',
        for val in item:
            print val,' | ',
        print ''
    for i in xrange(len(board)):
            print ' -----',

def valid(p,board,marker):
    p-=1
    if(p<0 or p>8):
        return False
    r = p/(len(board))
    c = p%(len(board))
    if(board[r][c]==' '):
        board[r][c]=marker
        return True
    else:
        return False

def row_check(r,board):
    x = board[r][0]
    if(x == ' '):
        return False
    for item in board[r]:
        if(item != x or item == ' '):
            return False
    return True
        
def col_check(c,board):
    x = board[0][c]
    if(x == ' '):
        return False
    for item in range(len(board)):
        if(board[item][c] != x or board[item][c] == ' '):
            return False
    return True

def l_diag_check(r,c,board):
    x = board[r][c]
    cnt=0
    l=r
    h=c
    while(l>=0 and h>=0):
        if(board[l][h] == x):
            cnt+=1
        else:
            return False
        l-=1
        h-=1
    
    l=r+1
    h=c+1
    while(l<len(board) and h<len(board)):
        if(board[l][h] == x):
            cnt+=1
        else:
            return False
        l+=1
        h+=1
    if(cnt==len(board)):
        return True
    else :
        return False

def r_diag_check(r,c,board):
    x = board[r][c]
    cnt=0
    l=r
    h=c
    while(l>=0 and h<len(board)):
        if(board[l][h] == x):
            cnt+=1
        else:
            return False
        l-=1
        h+=1
    
    l=r+1
    h=c-1
    while(l<len(board) and h>=0):
        if(board[l][h] == x):
            cnt+=1
        else:
            return False
        l+=1
        h-=1
    if(cnt==len(board)):
        return True
    else :
        return False

    
def win_check(p,board):
    p-=1
    r = p/(len(board))
    c = p%(len(board))
    if(row_check(r,board) or col_check(c,board) or l_diag_check(r,c,board) or r_diag_check(r,c,board)):
        return True
    else:
        return False

def tie_check(num,blen):
    if(num==blen):
        return True
    else:
        return False

board = [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
print 'Initial Board : '
print_board(board)
player_1 = raw_input("\nEnter the name of player 1 : ")
player_2 = raw_input("\nEnter the name of player 2 : ")
print "\n\n%s marker is 'X' and %s marker is 'O'\n\n"%(player_1,player_2)

blen=len(board)**2
num = 0

while(1):
    while(1):
        p1=int(raw_input("%s's turn , select a position on the board :"%(player_1)))
        if(valid(p1,board,'X')):
            break
        else :
            clear_output()
            print_board(board)
            print '\ninvalid input , please enter a valid cell '
            continue
    
    num+=1
    clear_output()
    print_board(board)
    
    if(win_check(p1,board)):
        print '\nGame over ! %s wins !!'%(player_1)
        break
    elif(tie_check(num,blen)):
        print '\nGame is a TIE !!'
        break
    
    while(1):
        p2=int(raw_input("%s's turn , select a position on the board :"%(player_2)))
        if(valid(p2,board,'O')):
            break
        else :
            clear_output()
            print_board(board)
            print '\ninvalid input , please enter a valid cell '
            continue
    
    num+=1
    clear_output()
    print_board(board)

    if(win_check(p2,board)):
        print '\nGame over ! %s wins !!'%(player_2)
        break
    elif(tie_check(num,blen)):
        print '\nGame is a TIE !!'
        break
    
```
